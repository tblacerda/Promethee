```python
import pandas as pd

import numpy as np
from pymcdm.methods import TOPSIS
from pymcdm.helpers import rrankdata
```

```python
pymcdm.methods.PROMETHEE_II
```

```python
df_ni = pd.read_excel('q2\Dataset Q2 - Projeto Ranking Multicriterio.xlsx')
```

```python
df_ni.head(5)
```

```python
weights = np.array(df_ni.iloc[0,9:].astype(float)) # SOMA 1
types = np.array(df_ni.iloc[1,9:].astype(float)) # 1 quanto maior melhor. 
                          #-1, quanto menor, menor.
                          
valores = np.array(df_ni.iloc[2:,9:]) # tirando a coluna com as OCs                    
```

```python
weights
```

```python
types
```

```python
# Create object of the method
topsis = TOPSIS()
```

```python
# Determine preferences and ranking for alternatives
pref = topsis(valores, weights, types)
ranking_ni = rrankdata(pref)
```

```python
df_ni = df_ni.iloc[2:,]
```

```python
df_ni['Ranking'] = ranking_ni
```

```python
df_ni.to_excel('q2/resultado.xlsx')
```
